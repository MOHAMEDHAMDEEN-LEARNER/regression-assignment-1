```python
x=df_copy.drop('FWI',axis=1)
y=df_copy['FWI']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train.corr()
```

```python
sns.heatmap(x_train.corr(),annot=True)
```

```python
def correlation(z,ther):
    col=set()
    corr_matrix=z.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>ther:
                colname=corr_matrix.columns[i]
                col.add(colname)
    return col
```

```python
cor_fea=correlation(x_train,0.85)
```

```python
cor_fea
```

```python
x_train.drop(cor_fea,axis=1,inplace=True)
x_test.drop(cor_fea,axis=1,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xs=scaler.fit_transform(x_train)
xs_t=scaler.transform(x_test)
```

```python
plt.subplot(1,2,1)
plt.boxplot(xs)
plt.subplot(1,2,2)
plt.boxplot(x_train)

plt.show()
```

# Linear Regression 

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(xs,y_train)
y_predict=reg.predict(xs_t)
plt.scatter(y_test,y_predict)
plt.plot(y_test,y_predict,'r')


```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
print(r2_score(y_test,y_predict))
```

### lasso Regression

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(xs,y_train)
y_predict=lasso.predict(xs_t)
plt.scatter(y_test,y_predict)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
print(r2_score(y_test,y_predict))
```

# lasso CV

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(xs,y_train)
y_predict=lassocv.predict(xs_t)
plt.scatter(y_test,y_predict)
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
print(r2_score(y_test,y_predict))
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(xs,y_train)
y_predict=ridge.predict(xs_t)
plt.scatter(y_test,y_predict)
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
print(r2_score(y_test,y_predict))
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(xs,y_train)
y_predict=ridgecv.predict(xs_t)
plt.scatter(y_test,y_predict)
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
print(r2_score(y_test,y_predict))
```

# Elastic Net

```python
from sklearn.linear_model import ElasticNet
Elastic=ElasticNet()
Elastic.fit(xs,y_train)
y_predict=Elastic.predict(xs_t)
plt.scatter(y_test,y_predict)
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
print(r2_score(y_test,y_predict))
```
